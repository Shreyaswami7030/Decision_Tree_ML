```python
df=sns.load_dataset("iris")
```

```python
df
```

```python
df.head()
```

```python
df["species"].replace({"setosa": 0,"versicolor": 1, "virginica":2},inplace=True)
```

```python
df["species"].value_counts().to_dict()
```

```python
# independent  feature and dependent feature

x= df.drop(["species"],axis=1)
y= df["species"]
```

```python
x
```

```python
y
```

```python
# train test split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
x_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Postpuruning
post_model= DecisionTreeClassifier(max_depth=2) 
```

```python
post_model.fit(x_train,y_train)
```

# Designing the decision tree

```python
from sklearn import tree
plt.figure(figsize=(10,10))
tree.plot_tree(post_model,filled=True)
```

```python
y_pred=post_model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score= accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```
